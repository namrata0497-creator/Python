## Basic Customer Profile Insights

```python
##Gender Counts - Identifying the total number of Male and Female Customers
gender_counts=df['Gender'].value_counts()
gender_counts
```

```python
##Gender Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df, palette='viridis')
plt.title('Gender Distribution')
plt.show()

gender_counts=df['Gender'].value_counts(normalize=True)*100
gender_counts
```

```python
##Geographic Distribution of Customers
geo_counts=df['Geography'].value_counts()
geo_counts
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Geography', data=df, palette='mako')
plt.title('Customer Distribution by Country')
plt.show()

geo_counts=df['Geography'].value_counts(normalize=True)*100
geo_counts
```

```python
##Age Distribution
plt.figure(figsize=(8,4))
sns.histplot(df['Age'], bins=20, kde=True, color='Teal')
plt.title('Age Distribution')
plt.show()

print('Average Age:', df['Age'].mean())
```

```python
##Tenure Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Tenure', data=df, palette='cool')
plt.title('Customer Tenure Distribution')
plt.show()

print('Average Tenure:', df['Tenure'].mean())
```

## Customer Financial Insights

```python
##Credit Score
plt.figure(figsize=(8,4))
sns.histplot(df['CreditScore'], bins=20, kde=True, color='purple')
plt.title('Credit Score Distribution')
plt.show()

print("Average_credit_score:", df['CreditScore'].mean())
```

```python
Num_of_active_customers=df['IsActiveMember'].value_counts()
Num_of_active_customers
```

```python
##Number of products owned
plt.figure(figsize=(8,4))
sns.histplot(x='NumOfProducts', data=df, palette='crest')
plt.title('Number of Products Owned')
plt.show()

print("Total_num_of_products_owned:", df['NumOfProducts'].value_counts())
```
